```python
import pandas as pd
import numpy as np
import lightgbm as lgb

filepath = 'F:/contest/kuaishou/'
app_path = 'app_launch_log.csv'
activity_path = 'user_activity_log.csv'
video_path = 'video_create_log.csv'
register_path = 'user_register_log.csv'

data_reg = pd.read_csv(filepath + register_path)
data_app = pd.read_csv(filepath + app_path)
data_acti = pd.read_csv(filepath + activity_path)
data_video = pd.read_csv(filepath + video_path)
```

```python
num_train = 3
train_reg = []
train_app = []
train_acti = []
train_video = []

for i in range(num_train):

    train_reg.append(data_reg[data_reg.date <= (16+i*7)])
    train_app.append(data_app[(data_app.date <= (16+i*7)) & (data_app.date >= (1+i*7))])
    train_acti.append(data_acti[(data_acti.date <= (16+i*7)) & (data_acti.date >= (1+i*7))])
    train_video.append(data_video[(data_video.date <= (16+i*7)) & (data_video.date >= (1+i*7))])
 
```

```python
data1 = pd.merge(train_reg[0],train_acti[0],on = 'id', how = 'left')
```

```python
#注册日志特征




#启动日志特征
def count_num(data):
    return data.groupby(['id']).size()

def gap_last(data, predict_day):
    tmp = data.groupby(['id'])
    gap = []
    last = []
    maxcontinue = []
    for i in tmp:
        gap.append(i[1].date.diff().mean())
        last.append(predict_day - max(i[1].date))
    return gap,last

    
# 行为日志特征
def count_page(data):
    tmp = data.groupby(['id'])
    page_0 = []
    page_1 = []
    page_2 = []
    page_3 = []
    page_4 = []
    for i in tmp:
        page_0.append((i[1].page == 0).sum())
        page_1.append((i[1].page == 1).sum())
        page_2.append((i[1].page == 2).sum())
        page_3.append((i[1].page == 3).sum())
        page_4.append((i[1].page == 4).sum())
    return page_0,page_1,page_2,page_3,page_4
def count_action(data):
    tmp = data.groupby(['id'])
    action_0 = []
    action_1 = []
    action_2 = []
    action_3 = []
    action_4 = []
    action_5 = []
    for i in tmp:
        action_0.append((i[1].action == 0).sum())
        action_1.append((i[1].action == 1).sum())
        action_2.append((i[1].action == 2).sum())
        action_3.append((i[1].action == 3).sum())
        action_4.append((i[1].action == 4).sum())
        action_5.append((i[1].action == 5).sum())
    return action_0,action_1,action_2,action_3,action_4,action_5
def maxnum_sameauthor(data):
    tmp = data.groupby(['date'])
    sameauthor = []
    samevideo = []
    for i in tmp:
        sameauthor.append(max(i[i].author.value_counts()))
        samevideo.append(max(i[i].video.value_counts()))
    return sameauthor,samevideo

#拍摄日志特征
```

```python
%time
for i in range(5):
    #注册日志
    dayofpredict = 20 + i
    feature[i] = train_reg[i]
    feature[i]['date'] = dayofpredict - feature[i]['date']
    feature[i] = pd.concat([feature[i], pd.get_dummies(feature[i].source, prefix = 'source')], axis = 1).drop(['source'], axis = 1)
        
    #启动日志
    feature[i] = pd.merge(feature[i],count_num(train_app[i]).rename('launch_num'), on = 'id', how = 'left')
    tmp1,tmp2 = gap_last(train_app[i],dayofpredict)
    feature[i] = pd.merge(feature[i],tmp1.rename('launch_gap'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('launch_last'), on = 'id', how = 'left')
    
   
    #行为日志
    feature[i] = pd.merge(feature[i],count_num(train_acti[i]).rename('acti_num'), on = 'id', how = 'left')
    tmp1 = feature[i]['acti_num']/feature[i]['launch_num']
    feature[i] = pd.merge(feature[i],tmp1.rename('acti_meandays'), on = 'id', how = 'left')
    tmp1,tmp2 = gap_last(train_acti[i],dayofpredict)
    feature[i] = pd.merge(feature[i],tmp1.rename('acti_gap'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('acti_last'), on = 'id', how = 'left')
    tmp1,tmp2,tmp3,tmp4,tmp5 = count_page(train_acti[i])
    feature[i] = pd.merge(feature[i],tmp1.rename('page_0'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('page_1'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp3.rename('page_2'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp4.rename('page_3'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp5.rename('page_4'), on = 'id', how = 'left')
    for j in range(5):
        feature[i]['page_'+str(j)+'_freq'] = feature[i]['page_'+str(i)]/feature[i]['acti_num'] if feature[i]['acti_num'].notnull() else 0
    tmp1,tmp2,tmp3,tmp4,tmp5,tmp6 = count_action(train_acti[i])
    feature[i] = pd.merge(feature[i],tmp1.rename('action_0'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('action_1'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp3.rename('action_2'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp4.rename('action_3'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp5.rename('action_4'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp6.rename('action_5'), on = 'id', how = 'left')
    for j in range(6):
        feature[i]['action_'+str(i)+'_freq'] = feature[i]['action_'+str(i)]/feature[i]['acti_num'] if feature[i]['acti_num'].notnull() else 0
    tmp1,tmp2 = maxnum_sameauthor(train_acti[i])
    feature[i] = pd.merge(feature[i],tmp1.rename('sameauthor'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('samevideo'), on = 'id', how = 'left')

    
    #拍摄日志
    feature[i] = pd.merge(feature[i],count_num(train_video[i]).rename('video_num'), on = 'id', how = 'left')
    tmp1,tmp2 = gap_last(train_video[i],dayofpredict)
    feature[i] = pd.merge(feature[i],tmp1.rename('video_gap'), on = 'id', how = 'left')
    feature[i] = pd.merge(feature[i],tmp2.rename('video_last'), on = 'id', how = 'left')
    
```
