```python
import lightgbm as lgb
LGBM = lgb.LGBMClassifier(
                         max_depth=7, 
                         n_estimators=60, 
                         learning_rate=0.1, 
                         num_leaves=63, 
                         min_samples_split=2)
df = data_train

X = df.iloc[:,1:50].as_matrix()
y = df.iloc[:,0].as_matrix()

LGBM.fit(X,y)
test = pd.read_csv("F:/contest/ATEC/data_test.csv")
df_test = test.iloc[:,0:49].as_matrix()
res = LGBM.predict(df_test)
```
